### we can see that PhoneService, StreamingTV and StreamingMovies are  contributing to a higher monthly charge and resulting to customers churning. Have tried to use for loop  as shown below

```python
services = ('PhoneService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection','MultipleLines', 'TechSupport', 'StreamingTV', 'StreamingMovies')

fig, ax = plt.subplots(figsize=(10, 6))

# Define the width of each bar
bar_width = 0.1

# Loop over each service and plot the bars side by side
for i, service in enumerate(services):
    a = data[data[service] == 'Yes']
    b = a.groupby('Churn').sum('MonthlyCharges').reset_index()
    x = np.arange(len(b.Churn))
    ax.bar(x + (i - 3) * bar_width, b.MonthlyCharges, width=bar_width, alpha=0.6, label=service)

# Set the x-axis ticks and tick labels
ax.set_xticks(np.arange(len(b.Churn)))
ax.set_xticklabels(b.Churn)

ax.set_title('Monthly Charges by Churn Status for Telecommunication Services')
ax.set_xlabel('Churn')
ax.set_ylabel('Monthly Charges')
ax.legend()
plt.show()
```

#  If the telco was to offer Phone as a standalone service, which type of contract  would encourage customer retention?


```python
data.PhoneService
```

```python
data.head()
```

```python


phone_data = data[data['PhoneService'] == 'Yes']
contract_data = phone_data.groupby('Contract')['Churn'].sum().reset_index()

# Create a bar chart of the contract types and their corresponding churn rates
plt.bar(contract_data['Contract'], contract_data['Churn'])
plt.xlabel('Contract Type')
plt.ylabel('Churn Rate')
plt.title('Churn Rates by Contract Type for Phone Services')
plt.show()
```

```python
data.head()
```

```python


# Filter for Phone Service customers
PhoneService_customers = data[data.PhoneService == 'Yes']

# Get unique contract types
PhoneService_customers.Contract.unique()

# Group by contract type and count the number of customers
PhoneService_customers.groupby('Contract').count().reset_index()

# Keep only relevant columns
PhoneService_customers = PhoneService_customers[['Contract', 'Churn']]

# Group by contract type and churn status, and count the number of customers
PhoneService_customers.groupby(['Contract', 'Churn']).size()

# Create a crosstab table
crosstab = pd.crosstab(PhoneService_customers['Contract'], PhoneService_customers['Churn'])

# Pivot the table to have Churn as columns
df = pd.DataFrame(crosstab).pivot(index='Contract', columns='Churn', values='count')

# Generate a stacked bar plot
ax = df.plot(kind='bar', stacked=True)
ax.set_xlabel('Contract')
ax.set_ylabel('Count')
ax.set_title('Customer Churn by Contract Type')

plt.show()
```

# b. Should the company strictly go paperless for their monthly billings as a technique of keeping their customers and why

```python
sns.countplot(x='PaperlessBilling', hue='Churn', data=data)
plt.title('Churn Distribution by Paperless Billing')
plt.xlabel('Paperless Billing')
plt.ylabel('Count')
plt.show()
```

```python
data['PaperlessBilling'] = data['PaperlessBilling'].replace({'Yes': 1, 'No': 0})

# Calculate churn rate for paperless billing customers
paperless_churn_rate = data.loc[data['PaperlessBilling'] == 1, 'Churn'].value_counts(normalize=True)[1]

print(f"Paperless billing churn rate: {paperless_churn_rate:.1%}")
```
