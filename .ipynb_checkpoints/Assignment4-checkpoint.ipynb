#  If the marketing team wanted to increase customer retention, to whom should they target their advertisements to?

```python
Churn = data['Churn'].value_counts()
Churn
```

```python
sns.catplot(data=data, x='gender', hue='Churn', kind='count')
plt.title('Telco Gender and Churn Distribution')
```

### To increase customer retention, the marketing team should target their advertisements to the customers who are most likely to churn. This can be determined by analyzing the customer data to identify the factors that contribute to churn, such as high monthly charges or low satisfaction with the services provided. By targeting these customers with personalized offers and promotions, the telco can improve customer retention and reduce churn.

#  Which 3 services are contributing to a higher monthly charge and resulting to customers churning?

```python
data.head()
```

```python
plot= sns.catplot(x="PaperlessBilling", y="MonthlyCharges", hue= "Churn", data=data, kind='bar')
# Set the x-axis label
plot.set(xlabel='PaperlessBilling', ylabel='Monthlycharges')

# Rotate the x-axis labels vertically
plot.set_xticklabels(rotation=90)
```

```python

sns.barplot(x='StreamingMovies', y='MonthlyCharges', hue='Churn', data=data, palette='muted')
plt.title('Monthly Charges  and Churn Status')
plt.xlabel('Service Category')
plt.ylabel(' Monthly Charges')
plt.show()
```

```python
sns.barplot(x='StreamingTV', y='MonthlyCharges', hue='Churn', data=data, palette='muted')
plt.title(' Monthly Charges  and Churn Status')
plt.xlabel('Service Category')
plt.ylabel(' Monthly Charges')
plt.show()
```
